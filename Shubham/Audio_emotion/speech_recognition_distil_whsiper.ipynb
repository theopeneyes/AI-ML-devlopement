## Whisper Small Model

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model_id = "distil-whisper/distil-small.en"

model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)
```

```python
processor = AutoProcessor.from_pretrained(model_id)

pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    max_new_tokens=512,
    torch_dtype=torch_dtype,
    device=device,
)
```

```python
%%time
sample = r"/content/drive/MyDrive/Data Science /Sentiment From Audio/Test_Audio_Files/Tom cruise angry dialouge.mp3"
result = pipe(sample)
print(result["text"])
```

```python
import torch
from transformers import AutoModelForSpeechSeq2Seq, AutoProcessor, pipeline
# from datasets import load_dataset
import time
```

```python
device = "cuda:0" if torch.cuda.is_available() else "cpu"
torch_dtype = torch.float16 if torch.cuda.is_available() else torch.float32

model_id = "distil-whisper/distil-medium.en"
```

```python
model = AutoModelForSpeechSeq2Seq.from_pretrained(
    model_id, torch_dtype=torch_dtype, low_cpu_mem_usage=True, use_safetensors=True
)
model.to(device)
```

```python
from transformers import GenerationConfig

MULTILINGUAL = True  # set True for multilingual models, False for English-only

if MULTILINGUAL:
    generation_config = GenerationConfig.from_pretrained(model)
else:
    generation_config = GenerationConfig.from_pretrained(model)

# generation_config.push_to_hub(model_id)
```

```python
processor = AutoProcessor.from_pretrained(model_id)

pipe = pipeline(
    "automatic-speech-recognition",
    model=model,
    tokenizer=processor.tokenizer,
    feature_extractor=processor.feature_extractor,
    max_new_tokens=512,
    chunk_length_s=15,
    batch_size=64,
    torch_dtype=torch_dtype,
    # return_timestamps =True,
    device=device,
)
```

### Trimming Audio file into chunks

```python
import librosa
import numpy as np
import soundfile as sf

y, sr = librosa.load(r"/content/drive/MyDrive/Data Science /Sentiment From Audio/Test_Audio_Files/sad_MONOLOGUE.mp3")

chunk_duration = 30  # Adjust duration as needed (in seconds)

num_chunks = int(np.ceil(len(y) / (sr * chunk_duration)))

chunks = np.array_split(y, num_chunks)

trimmed_chunks = []
for chunk in chunks:
    trimmed_chunk, _ = librosa.effects.trim(chunk, top_db=30)  # Adjust threshold as needed
    trimmed_chunks.append(trimmed_chunk)

for i, chunk in enumerate(trimmed_chunks):
    filename = f"chunk_{i}.wav"
    # Write out audio as 24bit PCM WAV
    sf.write(filename, chunk, sr)
```
